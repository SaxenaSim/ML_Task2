```python
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol='species',outputCol='species_index')
indexed = indexer.fit(df).transform(df)
indexed.show()
```

```python
import seaborn as sns

sns.pairplot(indexed.toPandas(), hue="species_index")
```

```python
import matplotlib.pyplot as plt 

for column in indexed.columns:
    if column!='species_index':
        sns.boxplot(x='species_index',y=column,data=indexed.toPandas())
        plt.title(f'Boxplot of {column}')
        plt.show()
```

```python
from pyspark.ml.feature import VectorAssembler

feature_columns = ['sepal_length','sepal_width','petal_length','petal_width']
assembler = VectorAssembler(inputCols=feature_columns,outputCol='features')
transformed_df = assembler.transform(indexed)
transformed_df.show()
```

```python
from pyspark.ml.classification import RandomForestClassifier

(train_data,test_data) = transformed_df.randomSplit([0.8,0.2],seed=123)
rf = RandomForestClassifier(labelCol='species_index',featuresCol='features',numTrees=10)
model = rf.fit(train_data)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

predictions = model.transform(test_data)
predictions.show()
evaluator = MulticlassClassificationEvaluator(labelCol='species_index',predictionCol='prediction',metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
print(accuracy)
```

```python
predictions.select("species_index", "prediction").show()
```

```python
from pyspark.mllib.evaluation import MulticlassMetrics

prediction_and_label = predictions.select("prediction", "species_index").rdd.map(lambda row: (row[0], float(row[1])))
metrics = MulticlassMetrics(prediction_and_label)
print(metrics.confusionMatrix().toArray())
```

```python
labels = predictions.select("species_index").distinct().rdd.map(lambda x: x[0]).collect()

precision = {}
for label in labels:
    precision[label] = metrics.precision(label)

for label, prec in precision.items():
    print(f"Precision for label {label}: {prec}")
```

```python
recall = {}
for label in labels:
    recall[label] = metrics.recall(label)

for label, rec in recall.items():
    print(f"Recall for label {label}: {rec}")
```
